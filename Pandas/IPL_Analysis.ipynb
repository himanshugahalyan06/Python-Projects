```python
# Combine both innings with team and score
innings = pd.concat([
    data[['team1', 'first_ings_score']].rename(columns={'team1': 'team', 'first_ings_score': 'score'}),
    data[['team2', 'second_ings_score']].rename(columns={'team2': 'team', 'second_ings_score': 'score'})
])

# Group by team and calculate average
avg_score = innings.groupby('team')['score'].mean()
print(avg_score)
```
