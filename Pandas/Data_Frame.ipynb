```python
## single row ---> dataframe_name.iloc[number_of_row]
## it give all details of a single row for given index row 
data.iloc[2]
```

```python
## mutiple row --> dataframe_name.iloc[range of the row]
## give first five rows

data.iloc[:5]
```
